#8.Grouping vs Crosstab vs Pivot-table

```python
insurance_data = pd.read_csv('/content/insurance.csv')
```

```python
insurance_data.head()
```

```python
insurance_data.groupby(by=['region','smoker'])['expenses'].mean()
```

```python
pd.pivot_table(data = insurance_data,values = 'expenses',index = 'smoker',columns = 'region',aggfunc='mean')
```

```python
pd.crosstab(index =insurance_data['smoker'] ,columns = insurance_data['region'],margins = True)
```
