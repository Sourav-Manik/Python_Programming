#1.Univariate Analysis

```python
from matplotlib import pyplot as plt
%matplotlib inline #to view the graph
```

image.png

```python
mtcars.head(2)
```

###1.1.Histogram for continous data

```python
plt.figure(figsize = (9,7))
plt.hist(x ='mpg' ,data = mtcars)
plt.title('MPG distribution')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

###1.2.Count plot - Discrete data

```python
mtcars['cyl'].value_counts()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (9,7))
sns.countplot(x='cyl',y=None,data = mtcars)
plt.title('Cyl distribution')
plt.show()
```

```python
sns.distplot(a=mtcars['mpg'],hist = False)
```

```python
plt.figure(figsize = (9,7))
plt.pie(x=mtcars['gear'].value_counts(),labels=[3,4,5],autopct = '%1.0f%%')
plt.show
```

```python
plt.boxplot(x= mtcars['mpg'])
plt.show()
```

#2.Bivariate Analysis

```python
mtcars.groupby(by = 'cyl')['mpg'].mean()
```

```python
sns.barplot(x = 'cyl',y='mpg',data = mtcars)
```

```python
mtcars.head(3)
```

```python
plt.scatter(x = 'hp', y = 'mpg',data = mtcars)
plt.show()
```

```python
plt.scatter(x = 'disp', y = 'mpg',data = mtcars)
plt.show()
```

```python
sns.boxplot(x = 'cyl',y='mpg',data =mtcars)
plt.show()
```

#3.Multivariate Analysis 

```python
plt.figure(figsize = (120,120))
sns.pairplot(data = mtcars)
plt.show()
```

```python
mtcars_corr = mtcars.corr().round(2)
```

```python
mtcars_corr
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(data = mtcars_corr,annot = True)
```
